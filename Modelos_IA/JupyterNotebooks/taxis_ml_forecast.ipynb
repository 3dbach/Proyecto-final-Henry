```python
import pandas as pd

# ... (cargar noviembre y taxi_zones)

# Añadir columnas 'hora_del_dia' y 'dia_de_la_semana' a noviembre
octubre['hora_del_dia'] = octubre['tpep_pickup_datetime'].dt.hour
octubre['dia_de_la_semana'] = octubre['tpep_pickup_datetime'].dt.dayofweek

# Fusión con la columna 'pulocationid'
octubre = pd.merge(octubre, taxi_zones[['LocationID', 'borough', 'zone']], left_on='pulocationid', right_on='LocationID', how='left')

# Renombrar las columnas para evitar conflictos
octubre = octubre.rename(columns={'borough': 'pickup_borough', 'zone': 'pickup_zone'})

# Eliminar la columna redundante
octubre = octubre.drop('LocationID', axis=1)

# Fusión con la columna 'dolocationid'
octubre = pd.merge(octubre, taxi_zones[['LocationID', 'borough', 'zone']], left_on='dolocationid', right_on='LocationID', how='left')

# Renombrar las columnas para evitar conflictos
octubre = octubre.rename(columns={'borough': 'dropoff_borough', 'zone': 'dropoff_zone'})

# Eliminar la columna redundante
octubre = octubre.drop('LocationID', axis=1)

# Eliminar filas con valores numéricos negativos
octubre = octubre[(octubre.select_dtypes(include='number') >= 0).all(axis=1)]

# Eliminar filas con valores numéricos demasiado altos
valor_umbral_superior = 350 # Puedes ajustar este valor según tus necesidades
octubre = octubre[((octubre.select_dtypes(include='number') <= valor_umbral_superior) & (octubre.select_dtypes(include='number') >= 0)).all(axis=1)]

# Eliminar filas con valores numéricos negativos
octubre = octubre[(octubre['trip_distance'] >= 0) & (octubre['fare_amount'] >= 0) & (octubre['total_amount'] >= 0)]

# Eliminar filas con valores numéricos demasiado altos
valor_umbral_superior = 30 # Puedes ajustar este valor según tus necesidades
octubre = octubre[octubre['trip_distance'] <= valor_umbral_superior]

octubre.dropna(inplace = True)

# Eliminar columnas duplicadas
octubre = octubre.loc[:,~octubre.columns.duplicated()]


# Guardar DataFrame resultante como archivo Parquet
octubre.to_parquet('octubre.parquet')
```
